This function is to plot the confusion matrix.


```python
def AI(prompt, model='gpt-3.5-turbo'):
    messages = [{'role':'user','content':prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,#this is the degree of randomness of the model's output
    )
    return response.choices[0].message['content']
```

```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

```python
import matplotlib.pyplot as plt

# Model accuracies
model_acc = {'LogReg': [0.8333333333333334, 0.8464285714285713], 
             'SVM': [0.8333333333333334, 0.8482142857142856], 
             'Tree': [0.94444444444, 0.875], 
             'KNN': [0.8333333333333334, 0.8482142857142858]}

# Get the best accuracy for each model
best_acc = {model: max(acc) for model, acc in model_acc.items()}

# Set the colors for accuracy and best score
colors = ['#1f77b4', '#ff7f0e']

# Create a bar chart
fig, ax = plt.subplots(figsize=(8, 6))
for i, (model, acc) in enumerate(model_acc.items()):
    # Highlight the bar with highest accuracy
    if max(acc) == best_acc[model]:
        ax.bar(model, max(acc), color=colors[1], alpha=0.8)
    else:
        ax.bar(model, max(acc), color=colors[0], alpha=0.8)
    # Add the accuracy values to the bars
    for j, val in enumerate(acc):
        ax.text(i, val+0.01, f'{val:.2f}', ha='center', fontsize=10)
        
# Set the title and axis labels
ax.set_title('Model Accuracies', fontsize=16)
ax.set_xlabel('Model', fontsize=14)
ax.set_ylabel('Accuracy', fontsize=14)

# Set the y-axis limits
ax.set_ylim([0.8, 0.9])

# Show the plot
plt.show()
```
